# Тема “Вычисления с помощью Numpy”


### Задание 1
Импортируйте библиотеку Numpy и дайте ей псевдоним np.
Создайте массив Numpy под названием a размером 5x2, то есть состоящий из 
5 строк и 2 столбцов. Первый столбец должен содержать числа 1, 2, 3, 3, 1, а второй - числа 6, 8, 11, 10, 7. Будем считать, что каждый столбец - это признак, а строка - наблюдение. Затем найдите среднее значение по каждому признаку, используя метод mean массива Numpy. Результат запишите в массив mean_a, в нем должно быть 2 элемента.



```python
import numpy as np
```

```python
a = np.array([[1,6],
              [2,8],
              [3,11],
              [3,10],
              [1,7]], dtype=int)
```

```python
mean_a = a.mean(axis=0)
```

```python
mean_a
```

### Задание 2
Вычислите массив a_centered, отняв от значений массива “а” средние значения соответствующих признаков, содержащиеся в массиве mean_a. Вычисление должно производиться в одно действие. Получившийся массив должен иметь размер 5x2.


```python
a_centered = a - mean_a
```

```python
a_centered
```

### Задание 3
Найдите скалярное произведение столбцов массива a_centered. В результате должна получиться величина a_centered_sp. Затем поделите a_centered_sp на N-1, где N - число наблюдений.


```python
a_centered_sp = np.dot(a_centered.T[0],a_centered.T[1].T)
```

```python
a_centered_sp
```

```python
N = a.shape[0]
```

```python
result = a_centered_sp / (N - 1)
```

```python
result
```

Затем создайте датафрейм book cо столбцами author_id, book_title и price, в которых соответственно содержатся данные:

[450, 300, 350, 500, 450, 370, 290].


```python
import pandas as pd
```

```python
authors = pd.DataFrame({
    "author_id": [1, 2, 3],
    "author_name": ['Тургенев', 'Чехов', 'Островский']
})
```

```python
authors
```

```python
books = pd.DataFrame({
    "author_id": [1, 1, 1, 2, 2, 3, 3],
    "book_title": ["Отцы и дети", "Рудин", "Дворянское гнездо", "Толстый и тонкий", "Дама с собачкой", "Гроза", "Таланты и поклонники"],
    "Price": [450, 300, 350, 500, 450, 370, 290]
})
```

```python
books
```

### Задание 2
Получите датафрейм authors_price, соединив датафреймы authors и books по полю author_id.

```python
authors_price = pd.merge(authors, books, on='author_id', how='outer')
```

```python
authors_price
```

### Задание 3
Создайте датафрейм top5, в котором содержатся строки из authors_price с пятью самыми дорогими книгами.

Количество уникальных значений для каждой характеристики:

```python
count_dict = {col: len(df[col].unique()) for col in df}
count_dict
```

Вывод количества уникальных значений для характеристик, количество уникальных значений которых менее unique_val

```python
unique_val = 10
d_frames = {}
for col in df:
    if df[col].nunique() < unique_val:
        vc = df[col].value_counts()
        d_frames[col] = pd.DataFrame({col + '_val':vc.index, col + '_count':vc.values})

for frame in d_frames.values():
    print(frame)
```
